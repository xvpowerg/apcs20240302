## 輸出
%s 字串
%d 整數
%f 浮點數

```python
n1 = 10
n2 = 20
result = n1 + n2
msg =  "%d + %d = %d"%(n1,n2,result)
print(msg)
```

```python
n1 = 10
n2 = 20
result = n1 + n2
msg = f"{n1} + {n2} = {result}"
print(msg)
```
