# 練習：字串練習
## 檢查使用者輸入的字串是否為迴文

```python
inStr = input("請輸入字串:")
reStr = inStr[::-1]
if inStr == reStr:
    print("是迴文")
else:
    print("不是迴文")
```

## 關係運算符號


```python
a = 3
b = 5
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#比文字的順序 左往右比 一個字一個字比
# 小寫 > 大寫 > 數字
#字典序
a = "Hello"
b = "hello"
print(b > a)
print(b < a)

a2 = "Ken"
b2 = "Vivin"
print(a2 == b2)
print(a2 != b2)
```
