```python
aqi = 25
# 0 ~50 良好
# 51 ~ 100 普通
# 101 ~150 敏感不健康

# 151 ~ 200 所有不健康
# 201 ~ 300 非常不健康
# 301 危險
if aqi >= 0 and aqi <= 50:
    print("良好")
elif 51 <= aqi <= 100:
    print("普通")
elif aqi >= 101 and aqi <= 150:
    print("敏感不健康")
elif 151 <= aqi <= 200:
    print("所有不健康")
elif  aqi >= 201 and aqi <= 300:
    print("非常不健康")
else:
     print("危險")
```

## Range 範圍

```python
a = range(10)
print(a)
print(list(a))
```

```python
a = range(2,6)
print(list(a))
```

```python
a = range(0,6,2)
print(list(a))
b = range(6,0,-2)
print(list(b))
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(5,10):
    print(i)
```

```python
for i in range(10,0,-2):
    print(i)
```
